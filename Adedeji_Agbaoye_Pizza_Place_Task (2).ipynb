## Joining all 4 datasets methodically

```python


#Outer-joining datasets 'orders' and 'order_details' based on our common column 'order_id in order to identify pizzas from inventory that was not on checkout (not bought)

merged_df_1= pd.merge(orders, order_details, on='order_id', how= 'outer')

#Outer-joining datasets 'merged_df_1 and 'pizzas' based on our common column 'pizza_id'
merged_df_2 = pd.merge(pizzas, merged_df_1, on='pizza_id', how='outer')

#Outer inner-joined dataset that includes 'pizza_types based on our common column 'pizza_type_id
df = pd.merge(pizza_types, merged_df_2, on='pizza_type_id', how='outer')
```

## Analysis from pivot table showing revenue and quantity sold from pizza Type Variants

```python

# Create a pivot table with name and size as index and Total Price and quantity as values
pizza_pivot = new_df.pivot_table(index=['name', 'size'], values=['Total', 'quantity'], aggfunc='sum')

# Renaming columns for clarity
pizza_pivot = pizza_pivot.rename(columns={'Total': 'Total Revenue', 'quantity': 'Quantity Sold'})

# Creating a new column that calls price per pizza
pizza_pivot['Price per Pizza'] = pizza_pivot['Total Revenue'] / pizza_pivot['Quantity Sold']

# Calculate percentage contribution
pizza_pivot['Percentage Contribution'] = round((pizza_pivot['Total Revenue'] / pizza_pivot.groupby(level=0)['Total Revenue'].transform('sum')) * 100,2)

# Resetitng index of pizza_pivot because I need to merged with the original dataset
pizza_pivot = pizza_pivot.reset_index()

# Deduplicate new_df based on 'name', 'pizza_id, and 'size', keeping the first occurrence - then I will drop duplicates because I need just one occurence of each pizza_id. if not, all my rows in new_df will return
new_df_deduplicated = new_df[['name', 'size', 'pizza_id']].drop_duplicates(subset=['name', 'size'], keep='first')

# Left-joining pizza_pivot with deduplicated new_df to add pizza_id
pizza_pivot = pd.merge(pizza_pivot, new_df_deduplicated, on=['name', 'size'], how='left')

# Setting the index back to 'name' and 'size'
pizza_pivot = pizza_pivot.set_index(['name', 'size'])

#Arranging it this way is easier to get insights from
desired_column_order = ['pizza_id','Price per Pizza', 'Quantity Sold', 'Total Revenue', 'Percentage Contribution']
pizza_pivot = pizza_pivot[desired_column_order]

# Displaying the pivot table
pizza_pivot
```

## Pivot Table for Correlation Calculations

```python

# Creating the pivot table
corr_pivot = pd.pivot_table(new_df, values=['price', 'quantity'], index='name', aggfunc=['sum', 'mean'])

# Access and display the relevant data
total_revenue = corr_pivot[('sum', 'price')] # Accessing total revenue (sum of price)
average_price = corr_pivot[('mean', 'price')] # Accessing average price (mean of price)
total_quantity = corr_pivot[('sum', 'quantity')]  # Accessing total quantity (sum of quantity)
# New DataFrame with desired columns
result_df = pd.DataFrame({
    'Total Revenue': total_revenue,
    'Average Price': average_price,
    'Total Quantity': total_quantity,

})

# Displaying the DataFrame sorted by Total Revenue
result_df.sort_values('Total Revenue', ascending=False)
```

## Correlation Calculations

```python
corr_pivot['Total Revenue'] = result_df['Total Revenue']
corr_pivot['Average Price'] = result_df['Average Price']
corr_pivot['Total Quantity'] = result_df['Total Quantity']
corr_pivot['Total Revenue Rank'] = result_df['Total Revenue'].rank(ascending=False).astype(int)
corr_pivot['Average Price Rank'] = result_df['Average Price'].rank(ascending=False).astype(int)
corr_pivot['Quantity Rank'] = result_df['Total Quantity'].rank(ascending=False).astype(int)

#See corr names as a column
corr_pivot_reset = corr_pivot.reset_index()

corr_pivot_reset[['name', 'Total Revenue', 'Total Revenue Rank', 'Average Price', 'Average Price Rank', 'Total Quantity', 'Quantity Rank']]
```

```python
#Using Spearman's Rank Coeffcient to calculate correlation index between two values

correlation_totalrev_avg_price = corr_pivot_reset[['Total Revenue', 'Average Price']].corr()
correlation_totalrev_quantity = corr_pivot_reset[['Total Revenue', 'Total Quantity']].corr()
```

```python

# Creating a new column to store the ingredient count by using the split string method to count the elements in the column with the base delimiter being (,) and counting each of the slices
new_df['ingredient_count'] = new_df['ingredients'].str.split(',').str.len()

# Grouping by ingredient count and then summing total revenue
ingredient_revenue = new_df.groupby('ingredient_count')['Total'].sum()

# Creating a DataFrame for the result
ingredient_revenue_df = pd.DataFrame({'Ingredient Count': ingredient_revenue.index, 'Total Revenue': ingredient_revenue.values})

# Display the DataFrame
print(ingredient_revenue_df)

# Plotting the relationship using a bar chart
plt.figure(figsize=(10, 6))
plt.bar(ingredient_revenue_df['Ingredient Count'], ingredient_revenue_df['Total Revenue'], color = 'darkred')
plt.xlabel('Number of Ingredients')
plt.ylabel('Total Revenue')
plt.title('Total Revenue vs. Number of Ingredients')
plt.show()
```

Pizzas cooked with 5 and 6 ingredients seem to be popular than others. It appears people like a balance - not too few ingredients, and not too much ingredients

```python
correlation_ingredient_revenue = ingredient_revenue_df.corr()
print("Correlation between ingredient number and revenue generated is: ", round(correlation_ingredient_revenue.iloc[0,1],2))
```

Well, the corrrelation of 0.37 is medium-weak but it might be worth pondering over because of the large sample size
