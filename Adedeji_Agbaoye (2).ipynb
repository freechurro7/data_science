## Joining all 4 datasets methodically

```python


#Outer-joining datasets 'orders' and 'order_details' based on our common column 'order_id in order to identify pizzas from inventory that was not on checkout (not bought)

merged_df_1= pd.merge(orders, order_details, on='order_id', how= 'outer')

#Outer-joining datasets 'merged_df_1 and 'pizzas' based on our common column 'pizza_id'
merged_df_2 = pd.merge(pizzas, merged_df_1, on='pizza_id', how='outer')

#Outer inner-joined dataset that includes 'pizza_types based on our common column 'pizza_type_id
df = pd.merge(pizza_types, merged_df_2, on='pizza_type_id', how='outer')
```

```python
df.tail(20)
```

```python
#Re-arranging the columns

new_df = df[['order_details_id', 'order_id', 'datetime', 'pizza_id', 'pizza_type_id', 'name', 'category', 'size', 'quantity', 'price', 'ingredients']]
```

```python
new_df.info()
```

## Cummulative Sales Progession by Day

```python

# Sort by datetime
new_df = new_df.sort_values(by='datetime')

# Calculating cumulative sales with cumsum()
new_df['Cumulative_Total'] = new_df['Total'].cumsum()

# Plot cumulative sales over time
plt.figure(figsize=(12, 6))
plt.plot(new_df['datetime'], new_df['Cumulative_Total'], color = 'green', marker='+')
plt.xlabel('Date')
plt.ylabel('Cumulative Sales')
plt.title('Cumulative Sales Over Time')
plt.grid(True)
plt.tight_layout()  #Subplots should fit into the figure area
plt.show()
```

## General pivot table for pizza names

```python

# Creating the pivot table
pizza_pivot = pd.pivot_table(new_df, values=['price', 'quantity'], index='name', aggfunc=['sum', 'mean'])

# Access and display the relevant data
total_revenue = pizza_pivot[('sum', 'price')] # Accessing total revenue (sum of price)
average_price = pizza_pivot[('mean', 'price')] # Accessing average price (mean of price)
total_quantity = pizza_pivot[('sum', 'quantity')]  # Accessing total quantity (sum of quantity)
# New DataFrame with desired columns
result_df = pd.DataFrame({
    'Total Revenue': total_revenue,
    'Average Price': average_price,
    'Total Quantity': total_quantity,

})

# Displaying the DataFrame sorted by Total Revenue
result_df.sort_values('Total Revenue', ascending=False)
```

## Correlation Calculations

```python
pizza_pivot['Total Revenue'] = result_df['Total Revenue']
pizza_pivot['Average Price'] = result_df['Average Price']
pizza_pivot['Total Quantity'] = result_df['Total Quantity']
pizza_pivot['Total Revenue Rank'] = result_df['Total Revenue'].rank(ascending=False).astype(int)
pizza_pivot['Average Price Rank'] = result_df['Average Price'].rank(ascending=False).astype(int)
pizza_pivot['Quantity Rank'] = result_df['Total Quantity'].rank(ascending=False).astype(int)

#See pizza names as a column
pizza_pivot_reset = pizza_pivot.reset_index()

pizza_pivot_reset[['name', 'Total Revenue', 'Total Revenue Rank', 'Average Price', 'Average Price Rank', 'Total Quantity', 'Quantity Rank']]
```

```python
#Using Spearman's Rank Coeffcient to calculate correlation index between two values

correlation_totalrev_avg_price = pizza_pivot_reset[['Total Revenue', 'Average Price']].corr()
correlation_totalrev_quantity = pizza_pivot_reset[['Total Revenue', 'Total Quantity']].corr()
```

```python
correlation_totalrev_avg_price
```

```python
correlation_totalrev_quantity
```

```python
#Retrives the output of calculated corr() value at row 0 and column 1
print("Correlation between Total Revenue and Average Price:", round(correlation_totalrev_avg_price.iloc[0, 1],3))
print("Correlation between Total Revenue and Quantity Sold:", round(correlation_totalrev_quantity.iloc[0, 1],3))
```

Correlation between total revenue and average price of 0.029 is a very weak postive correlation but that between total revenue and quantity sold (0.926) shows a very strong linear relationship
